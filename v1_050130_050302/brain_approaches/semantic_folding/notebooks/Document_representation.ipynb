```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Load your cleaned corpus from the "cleaned_corpus.txt" file
with open("/content/drive/MyDrive/cleaned_corpus.txt", "r") as file:
    corpus = file.readlines()

# Step 1: TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Step 2: Dimensionality Reduction using SVD
n_components = 400  # You can adjust the number of components as mentioned in your thesis
svd = TruncatedSVD(n_components=n_components)
document_vectors = svd.fit_transform(tfidf_matrix)

# Now, 'document_vectors' contains your document vectors with reduced dimensionality

# You can access the TF-IDF features and SVD components if needed:
# tfidf_features = tfidf_vectorizer.get_feature_names_out()
# svd_components = svd.components_
print(document_vectors)
# Example: Printing the top words associated with the first component of SVD
# top_words_idx = np.argsort(svd_components[0])[::-1][:10]
# top_words = [tfidf_features[idx] for idx in top_words_idx]

# print("Top words for the first SVD component:")
# print(top_words)
```
