```python
import numpy as np
import matplotlib.pyplot as plt
from minisom import MiniSom
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Load your cleaned corpus from the "cleaned_corpus.txt" file
with open("/content/drive/MyDrive/mock_corpus.txt", "r") as file:
    corpus = file.readlines()

# Step 1: TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=10)
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Step 2: Dimensionality Reduction using SVD
n_components = 10
svd = TruncatedSVD(n_components=n_components)
document_vectors = svd.fit_transform(tfidf_matrix)

# Step 3: Document Clustering with SOM
som_x, som_y = 30, 30
som = MiniSom(som_x, som_y, n_components, sigma=0.5, learning_rate=0.5)
som.pca_weights_init(document_vectors)
som.train_random(document_vectors, 1000)

# Step 4: Visualize the SOM Clusters with Document IDs
bmu_matrix = np.array([som.winner(d) for d in document_vectors]).T

# Plotting the SOM clusters with Document IDs
plt.figure(figsize=(som_x, som_y))
for i, vec in enumerate(document_vectors):
    document_id = i + 1  # Document ID starts from 1
    plt.text(bmu_matrix[0][i] + 0.5, bmu_matrix[1][i] + 0.5, f"ID:{document_id}", ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.5, lw=0))
plt.xticks(range(som_x))
plt.yticks(range(som_y))
plt.grid()
plt.show()
```
