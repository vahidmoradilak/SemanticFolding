```python
import spacy

# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')

# Define the input and output file paths
input_file_path = '/content/drive/MyDrive/openbook.txt'
output_file_path = '/content/drive/MyDrive/cleaned_corpus.txt'

# Read in the corpus text file and split it into lines
with open(input_file_path, 'r') as f:
    lines = f.readlines()

# Process and clean each line with spaCy and apply lemmatization
cleaned_lines = []
for line in lines:
    # Process the line with spaCy
    doc = nlp(line)

    # Filter tokens to remove stop words, punctuation, and non-alphabetic tokens
    filtered_tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]

    # Join the filtered tokens into a cleaned line
    cleaned_line = ' '.join(filtered_tokens)

    # Append the cleaned line to the list of cleaned lines
    cleaned_lines.append(cleaned_line)

# Write the cleaned lines to the output file
with open(output_file_path, 'w') as output_file:
    output_file.write('\n'.join(cleaned_lines))

# Print a message to confirm that the text has been written to the file
print(f'Cleaned and lemmatized text has been written to {output_file_path}')
```
